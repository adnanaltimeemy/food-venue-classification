```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create dataset dictionary
data = {
    "Store_ID": [f"NERO_{i:03d}" for i in range(1, 11)],
    "City": [
        "London", "Manchester", "Birmingham", "Leeds", "Glasgow",
        "Bristol", "Edinburgh", "Liverpool", "Sheffield", "Newcastle"
    ],
    "Annual_Sales_£m": [45.2, 26.5, 22.8, 20.4, 18.9, 17.3, 16.8, 15.7, 14.9, 13.5],
    "Profit_Margin_%": [12.5, 11.8, 11.2, 10.5, 10.0, 9.7, 9.5, 9.0, 8.8, 8.5],
    "Avg_Customer_Age": [34, 31, 30, 29, 33, 32, 35, 30, 28, 31],
    "Pct_Students": [25, 22, 20, 18, 15, 17, 12, 19, 18, 16],
    "Pct_Professionals": [55, 53, 50, 52, 58, 54, 60, 50, 48, 52]
}

# Load into DataFrame
df = pd.DataFrame(data)
df
```

```python
# Show first rows
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Annual_Sales_£m", y="Profit_Margin_%", hue="City", s=100)
plt.title("Sales vs Profit Margin per Café Nero Store")
plt.xlabel("Annual Sales (£m)")
plt.ylabel("Profit Margin (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
df[["Pct_Students", "Pct_Professionals"]].plot(kind="bar")
plt.title("Customer Segments per Store")
plt.xlabel("Store Index")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)
plt.show()
```
