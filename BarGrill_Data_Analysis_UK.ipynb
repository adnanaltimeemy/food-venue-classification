```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load sales data
sales_df = pd.read_csv("restaurant_sales_data.csv")

# Display first few rows
sales_df.head()
```

```python
# Convert Date & Time
sales_df['Date'] = pd.to_datetime(sales_df['Date'])
sales_df['Hour'] = sales_df['Date'].dt.hour

# Calculate Profit per Sale
sales_df['Profit'] = sales_df['Revenue'] - sales_df['Cost']

# Check for missing values
sales_df.isnull().sum()
```

```python
total_revenue = sales_df['Revenue'].sum()
print("Total Revenue:", total_revenue)
```

```python
hourly_transactions = sales_df.groupby('Hour').size()
hourly_transactions.plot(kind='bar', title='Transactions Per Hour')
```
