```python
import pandas as pd

# 1️⃣ Load without parsing dates first
df = pd.read_csv("manchester_pizza_sales.csv")

# 2️⃣ Check column names
print("Columns in dataset:")
print(df.columns)

# 3️⃣ If your column is called 'Date' (very common case)
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])
    df.set_index("Date", inplace=True)

# 4️⃣ If your column is called 'date' (lowercase)
elif "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"])
    df.set_index("date", inplace=True)

# 5️⃣ Show first rows
print("\nFirst 5 rows:")
print(df.head())
```
