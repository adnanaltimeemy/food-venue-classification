```python
# Step 1 — Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# Create sample data for American meals in UK
np.random.seed(42)

restaurants = ["Wendy's UK", "Slim Chickens", "Ed's Easy Diner", "Zaks Diners", "Dave's Hot Chicken"]
cities = ["London", "Manchester", "Birmingham", "Glasgow", "Leeds"]

age_groups = ["18-24", "25-34", "35-44", "45-54", "55+"]
genders = ["Male", "Female", "Non-Binary"]

data = []

for i in range(500):
    data.append({
        "Restaurant": np.random.choice(restaurants),
        "City": np.random.choice(cities),
        "Meal_Type": np.random.choice(["Burger", "BBQ", "Fried Chicken", "Southern Comfort"]),
        "Annual_Sales": np.random.randint(200000, 2000000),  # in GBP
        "Profit_Margin": round(np.random.uniform(3, 15), 1),  # restaurants typical 3-7% net, but margin varies
        "Age_Group": np.random.choice(age_groups, p=[0.2,0.3,0.25,0.15,0.1]),
        "Gender": np.random.choice(genders, p=[0.48,0.48,0.04]),
        "Visits_per_Week": np.random.randint(1, 8),
    })

df = pd.DataFrame(data)
df.head()
```

```python
# Basic summary stats
df.describe()
```

```python

plt.figure(figsize=(10,6))
sns.barplot(x="Meal_Type", y="Annual_Sales", data=df, estimator=np.mean, ci=None)
plt.title("Average Annual Sales by American Meal Type (UK)")
plt.ylabel("Avg Sales (£)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Meal_Type", y="Profit_Margin", data=df)
plt.title("Profit Margin by Meal Type")
plt.ylabel("Profit Margin (%)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Age_Group", hue="Meal_Type", data=df)
plt.title("American Meal Preference by Age Group (UK)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Gender", data=df)
plt.title("Customer Gender Breakdown")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Annual_Sales", y="Profit_Margin", hue="Meal_Type", data=df)
plt.title("Sales vs Profit Margin by Meal Type")
plt.xlabel("Annual Sales (£)")
plt.ylabel("Profit Margin (%)")
plt.show()
```
