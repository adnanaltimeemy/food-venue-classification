```python
#  Miami Crispy Step  Sales, Profit & Demographics Analysis

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1) Create / Load Dataset ---

# Example: sales by month (hypothetical)
data = {
    "Month": pd.date_range(start="2025-01-01", periods=12, freq="M"),
    "Store_Location": ["Miami Beach"]*12,
    "Sales_USD": [12000,15000,18000,17500,16000,20000,22000,24000,23000,21000,19500,25000],
    "Profit_Margin_pct": [15,17,19,18,18.5,20,21,22,21.5,20,19.8,23]
}

# Demo demographic data  age groups of customers
demo_data = {
    "Age_Group": ["18-24","25-34","35-44","45-54","55+"],
    "Purchase_Count": [1200,2400,1900,1300,900],
    "Avg_Spend_USD": [15.5,18.2,20.1,22.5,17.8]
}

sales_df = pd.DataFrame(data)
demo_df = pd.DataFrame(demo_data)

# --- 2) Inspect Datasets ---
print("--- Sales Data ---")
display(sales_df.head())

print("--- Demographics ---")
display(demo_df)

# --- 3) Summary Statistics ---
print("Sales summary:")
print(sales_df["Sales_USD"].describe())

print("Profit margins summary:")
print(sales_df["Profit_Margin_pct"].describe())

# --- 4) Sales Trend Visualization ---
plt.figure(figsize=(10, 5))
sns.lineplot(x="Month", y="Sales_USD", data=sales_df, marker="o")
plt.title("Monthly Sales (Miami Crispy Step)")
plt.ylabel("Sales (USD)")
plt.tight_layout()
plt.show()

# Profit margin trend
plt.figure(figsize=(10, 5))
sns.lineplot(x="Month", y="Profit_Margin_pct", data=sales_df, marker="o", color="orange")
plt.title("Profit Margin % Over Time")
plt.ylabel("Profit Margin (%)")
plt.tight_layout()
plt.show()

# --- 5) Demographics Analysis ---

# Bar chart: Age Group vs Purchase Count
plt.figure(figsize=(8,5))
sns.barplot(x="Age_Group", y="Purchase_Count", data=demo_df, palette="magma")
plt.title("Customer Purchases by Age Group")
plt.ylabel("Purchases")
plt.xlabel("Age Group")
plt.tight_layout()
plt.show()

# Avg Spend by Demographic
plt.figure(figsize=(8,5))
sns.barplot(x="Age_Group", y="Avg_Spend_USD", data=demo_df, palette="viridis")
plt.title("Average Spend per Customer by Age Group")
plt.ylabel("Avg Spend (USD)")
plt.tight_layout()
plt.show()

# --- 6) Optional: Profit Projection ---
# Project future profit margin growth
sales_df["Projected_Profit"] = sales_df["Sales_USD"] * (sales_df["Profit_Margin_pct"] / 100)

print("Projected Profit figures:")
display(sales_df[["Month","Sales_USD","Projected_Profit"]])
```
